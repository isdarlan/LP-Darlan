```python
Geral =["início"]
Prioridade =["início"]

def menu():
  print(f"\nmenu principal")
  print(f"1 para adicionar cliente:")
  print(f"2 para atemder o cliente:")
  print(f"3 para ver as filas:")
  print(f"4 para relatorio das listas:")
  print(f"5 para sair do do menu")
  volta = int(input(f" digite aqui:"))
  if  volta == 1:
    adicionar()
  elif volta == 2:
    atender()
  elif volta == 3:
    verlista()
  elif volta == 4:
    relatório()
  elif volta == 5:
    print(f"obrigado por usar nosso sistema!")
  else:
    print("Opção inválida. Por favor, tente novamente.")
    menu()


def adicionar():
  print("\naqui é para adicionar o cliente:")
  nome = str(input(" escreva seu nome aqui:"))
  idade = int(input(" escreva sua idade aqui:"))
  gestante = input("você é gestante (s/n): ").lower() == 's'
  pcd = input("você é PCD (s/n): ").lower() == 's'

  if idade >= 60 or gestante or pcd:
    Prioridade.append(nome)
    numero = Prioridade.index(nome)
    print(f"Essa é sua posição na fila de prioridade: {numero}\n{Prioridade}")
  else:
    Geral.append(nome)
    numero = Geral.index(nome)
    print(f"Essa é sua posição na fila geral: {numero}\n{Geral}")

  resp = int(input("quer voltar para o menu:\n1 - para voltar\n2 - para sair\ndigite aqui:"))
  if resp == 1:
    menu()
  else :
    print("obrigado por usar nosso sistema!")


def atender():
  print("\nmenu de atendimento ao cliente:")
  print(f"1 para atender a fila geral:\n{Geral}")
  print(f"2 para atender a fila de prioridade:\n{Prioridade}")
  print("3 para voltar ao menu")
  volta = int(input("aqui:"))
  if volta == 1:
    ret = int(input("qual o numero dessa pessoa:"))
    if 0 <= ret < len(Geral):
      Geral.pop(ret)
      print(f" essa é a nova fila geral:\n{Geral}")
    else:
      print("Número inválido.")
    resp = int(input("quer voltar para o menu:\n1 - para voltar\n2 - para sair\ndigite aqui:"))
    if resp == 1:
      menu()
    else :
      print("obrigado por usar nosso sistema!")
  elif volta == 2:
    ret = int(input("qual o numero dessa pessoa:"))
    if 0 <= ret < len(Prioridade):
      Prioridade.pop(ret)
      print(f" essa é a nova fila de prioridade:\n{Prioridade}")
    else:
      print("Número inválido.")
    resp = int(input("quer voltar para o menu:\n1 - para voltar\n2 - para sair\ndigite aqui:"))
    if resp == 1:
      menu()
    else :
      print("obrigado por usar nosso sistema!")
  elif volta == 3:
    menu()
  else :
    print("Opção inválida. Por favor, tente novamente.")
    atender()

def verlista():
  print("\nessa opção só serve para ver as listas: ")
  print(f"essa é a fila geral:\n{Geral}")
  print(f"essa é a fila de prioridade:\n{Prioridade}")
  resp = int(input("quer voltar para o menu:\n1 - para voltar\n2 - para sair\ndigite aqui:"))
  if resp == 1:
    menu()
  else :
    print("obrigado por usar nosso sistema!")

def relatório():
  print(f"\nqual relatorio deve ser imprimido:")
  print("1 para relatorio geral:")
  print("2 para relatorio de prioridade:")
  print("3 para voltar ao menu")
  volta = int(input("aqui:"))

  if volta == 1:
    relatgeral()

  elif volta == 2:
    relatprioridade()

  elif volta == 3:
    menu()
  else:
    print("Opção inválida. Por favor, tente novamente.")
    relatório()


def relatgeral():
  lista_antiga = ["início"]
  set_antiga = set(lista_antiga)
  set_nova = set(Geral)

  itens_removidos = set_antiga - set_nova
  itens_adicionados = set_nova - set_antiga

  porcentagem_removidos = (len(itens_removidos) / len(lista_antiga)) * 100 if len(lista_antiga) > 0 else 0
  porcentagem_adicionados = (len(itens_adicionados) / len(lista_antiga)) * 100 if len(lista_antiga) > 0 else 0

  print("relatorio da fila geral")
  print(f"-total original: {len(lista_antiga)} itens")
  print(f"-itens removidos: {len(itens_removidos)} ({porcentagem_removidos:.2f}%)")
  print(f"-total atual: {len(Geral)} itens")
  print(f"-itens adicionados: {len(itens_adicionados)} ({porcentagem_adicionados:.2f}%)")

  resp = int(input("quer voltar para o menu:\n1 - para voltar\n2 - para sair\ndigite aqui:"))
  if resp == 1:
    menu()
  else :
    print("obrigado por usar nosso sistema!")


def relatprioridade():
  lista_antiga = ["início"]
  set_antiga = set(lista_antiga)
  set_nova = set(Prioridade)

  itens_removidos = set_antiga - set_nova
  itens_adicionados = set_nova - set_antiga

  porcentagem_removidos = (len(itens_removidos) / len(lista_antiga)) * 100 if len(lista_antiga) > 0 else 0
  porcentagem_adicionados = (len(itens_adicionados) / len(lista_antiga)) * 100 if len(lista_antiga) > 0 else 0

  print("relatorio da fila de prioridade")
  print(f"-total original: {len(lista_antiga)} itens")
  print(f"-itens removidos: {len(itens_removidos)} ({porcentagem_removidos:.2f}%)")
  print(f"-total atual: {len(Prioridade)} itens")
  print(f"-itens adicionados: {len(itens_adicionados)} ({porcentagem_adicionados:.2f}%)")

  resp = int(input("quer voltar para o menu:\n1 - para voltar\n2 - para sair\ndigite aqui:"))
  if resp == 1:
    menu()
  else :
    print("obrigado por usar nosso sistema!")

menu()
```
